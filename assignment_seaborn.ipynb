Step 5. Create a plot to check the relationship between horsepower and acceleration. Note down your insight for the same. Beautify the graph using various customizations.

```python
mpg_df.plot( x = 'horsepower', y = 'acceleration',kind = 'scatter')
plt.title('The Relationship between horsepower and acceleration',color='blue',fontsize=18)
plt.ylabel('acceleration',color='green',fontsize=15)
plt.xlabel('horsepower',color='green',fontsize=15)
plt.show()
```

Step 6. Generate subplots to display the histograms for acceleration, displacement and weight and kilometer_per_litre. Note down your insight for the same

```python
### In data set no ,kilometer_per_litre' column so in place 'kilometer_per_litre' i take 'mpg'
x = np.array(mpg_df['acceleration'])
y = np.array(mpg_df['displacement'])
z = np.array(mpg_df['mpg'])

#plot 1:
plt.subplot(1, 3, 1)
plt.hist(x)
plt.xlabel('acceleration',color='r',fontsize=20)

#plot 2:
plt.subplot(1, 3, 2)
plt.hist(y)
plt.xlabel('displacement',color='r',fontsize=20)

#plot 3:
plt.subplot(1, 3, 3)
plt.hist(z)
plt.xlabel('mpg',color='r',fontsize=20)

plt.show()
```
